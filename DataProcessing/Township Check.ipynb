```python
import numpy as np
import pandas as pd
import json
```

```python
d = None
with open("admin_level_8.json") as json_data:
    d = json.load(json_data)
```

```python
townships = []
```

```python
for f in d["features"]:
    if "name:en" in f["properties"]:
        townships.append(f["properties"]["name:en"])
```

```python
len(townships)
litData = pd.read_csv("literacy_myanmar.csv")
full_townships = litData.name_ts
```

```python
for t in full_townships:
    if t not in townships:
        print(t)
```

```python
for t in townships:
    print(t)
```

```python
# Only Mogoke Left
for f in d["features"]:
    if "name:en" not in f["properties"]:
        f["properties"]["name:en"] = "Mogoke"
```

```python
with open('admin_level_8_final.json', 'w') as fp:
    json.dump(d, fp)
```
